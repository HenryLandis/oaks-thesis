```python
import ipyrad.analysis as ipa
import toyplot
import toytree
```

Infer topologies for the remaining chromosomes from the assembly.

```python
wex01 = ipa.window_extracter(
    name = "basic_scaffold01",
    data = "/moto/eaton/users/hnl2109/analysis-ipyrad/robur_ref_basic_outfiles/robur_ref_basic.seqs.hdf5",
    workdir = "/moto/eaton/users/hnl2109/analysis-window_extracter",
    scaffold_idxs = [1],
    mincov = 0.5,
    rmincov = 0.1
)

wex01.stats
```

```python
wex01.run(force = False, nexus = False)
```

```python
rax01 = ipa.raxml(
    name = "basic_scaffold01",
    data = wex01.outfile,
    workdir = "/moto/eaton/users/hnl2109/analysis-raxml",
    N = 100,
    T = 12,
)

rax01.command
```

```python
rax01.run()
```

```python
tre01 = toytree.tree("/moto/eaton/users/hnl2109/analysis-raxml/RAxML_bipartitions.basic_scaffold01")
tre01.draw(width = 500, height = 3000, node_labels="support", use_edge_lengths=False);
```
