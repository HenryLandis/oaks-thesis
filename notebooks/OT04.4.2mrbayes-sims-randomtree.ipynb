```python
import toytree
import ipcoal
import ipyrad.analysis as ipa
import random
```

```python
trer = toytree.rtree.rtree(200, treeheight = 10e6, seed = 123)
trer.draw(layout = "d");
```

```python
modr = ipcoal.Model(trer, Ne = 1e5, nsamples = 2, seed = 123)
modr.sim_loci(1, 10)
```

```python
modr.write_concat_to_nexus(name="mbtest-r", outdir='/moto/eaton/users/hnl2109/analysis-mrbayes/', diploid=True)
```

```python
mbr2 = ipa.mrbayes(
    data="/moto/eaton/users/hnl2109/analysis-mrbayes/mbtest-r.nex",
    name="itest-r2",
    workdir="/moto/eaton/users/hnl2109/analysis-mrbayes/",
    clock_model=2,
    constraints=trer,
    ngen=int(1e6),
    nruns=2,
)

# Print parameters.
print(mbr2.params)
```

```python
mbr2.run()
```

```python
mbr2tre = toytree.tree("/moto/eaton/users/hnl2109/analysis-mrbayes/itest-r2.nex.con.tre", 10)
mbr2tre = mbr2tre.mod.node_scale_root_height(10e6)
c, a, m = mbr2tre.draw(ts='o', layout='d', scalebar=True);

# Overlay true tree.
trer.draw(
    axes=a, 
    ts='o', layout='d', scalebar=True, 
    edge_colors="darkorange",
    node_sizes=0,
);
```

Tree with variation in Ne
=====

```python
trerv = toytree.rtree.rtree(200, treeheight = 5.6e7, seed = 234)
trerv.draw(layout = "d", scalebar = True);
```

```python
# Create a dictionary of Ne values, chosen randomly from a given interval, then set onto the tree.
dict_ne = {i.name : random.randint(1e4, 2e4) for i in trerv.get_feature_dict()}
trerv = trerv.set_node_values("Ne", dict_ne)
```

```python
# We will assume gentime = 10 (that is, # of gens per year = 0.1) across the tree to start.
dict_g = {i.name : 10 for i in trerv.get_feature_dict()}
trerv = trerv.set_node_values("g", dict_g)
```

```python
# Divide edge lengths (absolute time) by generation time to get units of generations.
ttrerv = trerv.set_node_values(
    "dist",
    {i.name: i.dist / i.g for i in trerv.get_feature_dict()}
)
```

```python
ttrerv.draw(layout = "d", scalebar = True);
```

```python
modrv = ipcoal.Model(ttrerv, nsamples = 2)
modrv.sim_loci(1, 1000)
```

```python
modrv.write_concat_to_nexus(name="mbtest-rv", outdir='/moto/eaton/users/hnl2109/analysis-mrbayes/', diploid=True)
```

```python
modrv.df
```

```python
mbrv2 = ipa.mrbayes(
    data="/moto/eaton/users/hnl2109/analysis-mrbayes/mbtest-rv.nex",
    name="itest-rv2",
    workdir="/moto/eaton/users/hnl2109/analysis-mrbayes/",
    clock_model=2,
    constraints=ttrerv,
    ngen=int(1e6),
    nruns=2,
)

# Print parameters.
print(mbrv2.params)
```

```python
mbrv2.run(force = True)
```

```python
mbrv2tre = toytree.tree("/moto/eaton/users/hnl2109/analysis-mrbayes/itest-rv2.nex.con.tre", 10)
mbrv2tre = mbrv2tre.mod.node_scale_root_height(5.6e6)
c, a, m = mbrv2tre.draw(ts='o', layout='d', scalebar=True);

# Overlay true tree.
ttrerv.draw(
    axes=a, 
    ts='o', layout='d', scalebar=True, 
    edge_colors="darkorange",
    node_sizes=0,
);
```
